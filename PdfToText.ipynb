# Convertisser une PDF en text  avec PyPDF2 

```python
pip install PyPDF2
```

```python

from PyPDF2 import PdfReader

# Ouvrir le fichier PDF en mode lecture binaire
with open('SOA.pdf', 'rb') as file:
    # Instancier l'objet PdfFileReader
    pdf_reader = PdfReader(file)

    # Parcourir toutes les pages du PDF
    for page_number in range(len(pdf_reader.pages)):
        # Récupérer la page
        page = pdf_reader.pages[page_number]
        # Extraire le texte de la page
        text = page.extract_text()
        # Afficher le texte
        print(text)

        # Écrire le texte dans un fichier
        with open('SOA.txt', 'a') as text_file:
            text_file.write(text)
```
